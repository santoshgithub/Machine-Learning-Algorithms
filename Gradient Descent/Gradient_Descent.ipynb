```python
#Gradient descent scratch implementation
import numpy as np
```

```python
X = np.random.randn(20,1)
y = 3*X + np.random.rand()
```

```python
def gradient_descent(x, y, learning_rate=0.001, iterations=1000):
    # Initialize the parameters
    w = 0.0
    b = 0.0
    N = x.shape[0]
    # Iterate over the number of iterations
    for i in range(iterations):
        # Calculate the predictions
        y_pred = w * x + b
        
        # Calculate the loss
        loss = np.divide(np.sum((y-y_pred)**2, axis=0), x.shape[0]) 
        
        # Calculate the gradients
        dldw = np.sum(-2 * x * (y - y_pred))/N
        dldb = np.sum(-2 * (y - y_pred))/N
        
        # Update the parameters
        w = w - learning_rate * dldw
        b = b - learning_rate * dldb
        
        print(f'{i} loss is {loss}, paramters w:{w}, b:{b}')
    
    return w, b
```

```python
w, b = gradient_descent(X, y, learning_rate=0.01, iterations=1000)
```
